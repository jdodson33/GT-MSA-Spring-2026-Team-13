# Executive Exploratory Data Analysis

## Project Purpose Overview

Bitcoin’s volatility creates both risk and opportunity for accumulation strategies. Traditional approaches like Uniform DCA are simple but ignore market conditions. More sophisticated methods, such as Value Averaging or Dynamic DCA, adjust contributions based on indicators or targets. As your slides note, Dynamic DCA requires “market indicator weights” and “on‑chain/price data” to adapt allocations in real time. This project aims to design an original dynamic DCA model that leverages both CoinMetrics BTC data and Polymarket prediction‑market data to identify signals that may improve long‑run accumulation outcomes.


## Findings from Polymarket EDA
1. Odds Behavior Shows No Predictive Power

    A key empirical finding is that Polymarket odds for crypto‑related markets are overwhelmingly concentrated around 0.5, indicating that most markets never meaningfully update their probabilities. The odds distribution is “highly centralized around $0.5,” with a mean of 0.500008 and extremely low variance.

    This suggests that Polymarket odds do not encode meaningful information about Bitcoin price movements.

2. Volume Correlations Are Weak

    Polymarket trading volume evaluated against:
    - 7‑day BTC volatility (stdev)
    - 1‑day, 5‑day, and 10‑day forward BTC returns
    
    No strong relationships emerged. For example, volume’s correlation with 7‑day volatility is –0.092, implying “a loosely inverse correlation.” Forward‑return correlations were similarly negligible.

    Conclusion: Polymarket data, at least in raw form, does not provide standalone predictive signals for BTC returns or volatility.


## Findings from Bitcoin Technical Indicator EDA

Computed a suite of classical technical indicators, including:

- SMA‑20
- MACD (12/26 EMA)
- RSI (14‑day)
- Stochastic Oscillator (14‑day)
- Distance to SMA‑20

Evaluated correlations between these indicators and:

- Daily BTC returns
- 1‑day, 5‑day, and 10‑day forward returns


1. Distance to SMA‑20 Shows the Strongest Signal

    The most notable relationship is that distance to SMA‑20 has a 0.246 correlation with 10‑day forward returns, the strongest signal identified so far. While modest, this suggests that mean‑reversion or momentum effects around short‑term moving averages may be exploitable in a dynamic DCA framework.

2. Net USD Flow Is Minimally Predictive

    Evaluated whether net USD inflows into BTC predict returns. The conclusion is clear: net flow is not strongly predictive of BTC returns.


## Conclusion and Next Steps

The exploratory analysis reveals that no single indicator—whether from Polymarket or Bitcoin technicals—provides strong predictive power in isolation. However, this does not preclude the possibility that combinations of weak signals may yield a meaningful composite predictor. Thus, next steps include the following action items:


1. Build Composite Signal Models

    Test combinations of:
    - Technical indicators (e.g., SMA distance, MACD, RSI)
    - Polymarket volume
    - Net USD flow
    - Other features found in the dataset

    The goal is to identify multi‑signal regimes where expected forward returns differ meaningfully from baseline.

2. Test Increased Focus on Market Decline / Volatility Increase

    Focus more directly on evaluating indicators of price drops or hightened volatility as opposed to BTC returns

3. Backtest Combination Strategies and Refine

    Test each signal combination with a backtest to refine a finalized model outperforming Uniform DCA
